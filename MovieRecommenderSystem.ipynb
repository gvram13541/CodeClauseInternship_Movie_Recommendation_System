# Content Based Recommender System

```python
new_df = df.head(10000) 
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

def recommend_videos(search_query):
    new_df_copy = new_df.copy()  # Create a deep copy of the DataFrame
    new_df_copy['content'] = new_df_copy['genre'] + ' ' + new_df_copy['name'] + ' ' + new_df_copy['description'] + ' ' + new_df_copy['year'].astype(str) + ' ' + new_df_copy['rating'].astype(str) + ' ' + new_df_copy['directors_name'] + ' ' + new_df_copy['stars_name'] + new_df_copy['certificate']
    new_df_copy['content'] = new_df_copy['content'].str.lower()

    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(new_df_copy['content'])

    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

    query_vector = tfidf_vectorizer.transform([search_query])
    cosine_scores = linear_kernel(query_vector, tfidf_matrix)

    video_indices = cosine_scores.argsort()[0][::-1]

    recommended_videos = new_df_copy.iloc[video_indices[:10]][['name']].values.tolist()

    return recommended_videos

# Example usage:
search_query = input("Enter the genre, movie name, description, year, rating, director name, star name, of the movies you are searching for: ")
recommended_videos = recommend_videos(search_query)

if recommended_videos:
    for name in recommended_videos:
        print(f"Title: {name[0]}")
else:
    print("No videos found matching your query. Please try a different search.")
```
